# Tips for choosing an appropriate visualization:

## Use Appropriate "Visual Vocabulary"

[Visual Vocabulary - Vega Edition](http://ft.com/vocabulary)

## What are the properties of your data?
- Is your primary variable of interest continuous or discrete?
- Is in wide or long (tidy) format?
- Does your visualization involve multiple variables?
- How many dimensions do you need to include on your plot?

Can you express the main idea of your visualization in a single sentence?

How hard does your visualization make the user work in order to draw the intended conclusion?

## Make prototypes

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#Use the fivethrityeight style from plt
plt.style.use('fivethirtyeight')
fig = plt.figure()

fake = pd.Series([38, 3, 2, 1, 2, 4, 6, 5, 5, 33], 
                 index=range(1,11))

#Set the color of the bars  and their width
ax = fake.plot.bar(color='#ef7030', width=0.9)

#set the spacing and number of increments on the y axis
#set the background color of the chart area
ax.set(yticks=range(0,50,10), facecolor='#fcfcfc')

#set the background color of the border 
fig.patch.set_facecolor('#fcfcfc')
#Label the axes and set to bold
plt.ylabel('Percent of total votes', fontsize=9, fontweight='bold')
plt.xlabel('Rating', fontsize=9, fontweight='bold')

#Set the title and give the coordinates for placing the title
ax.text(x=-2, y=46, s="'An Inconvenient Sequel: Truth To Power' is divisive",
       fontsize=12, fontweight='bold')

#Set the sub-title and give the coordinates for placing the title
ax.text(x=-2, y=43, s= 'IMDb ratings for the film as of Aug. 29',
       fontsize=11)

#The numbers on the x-axis were original on their sides. This rotates them
#upright
ax.tick_params(labelrotation=0)
```

## Reproduce with real data

```python
df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/inconvenient-sequel/ratings.csv')
```

```python
df.head()
```

```python
#This displays a random row and transposes it
df.sample(1).T
```

```python
#Just lists out the columns and their data types. Take a look to see if it makes sense to 
# cast any of them to a different type. In this case it does. Cast timestamp to datetime
# so the data can be more useful
df.dtypes
```

```python
#Changing the data type of the timestamp column 
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['timestamp'].describe()
```

```python
#Setting the DF index to the timestamp column and then looking at the rows for August 29th
#This is the last date in the data. The reason we did this was to see if the data is cumulative
# It is so we can just use that timestamp to get the cumulative numbers for the whole time period

df.set_index('timestamp', inplace=True)
df['2017-08-29']
```

```python
#just putting this in a separate cell because re-running set_index is bad news

df['2017-08-29']
```

```python
#DF with just the August 29th rows
lastday = df['2017-08-29']

#Further iltering the lastday df to grab the IMDb users which should give us the cumulative
# numbers for all users which is what we need to recreate the 538 chart 
lastday_filtered = lastday[lastday['category'] == 'IMDb users']


```

```python
#Grab the very last row in the lastday_filtered DF and then transpose
# to make it all tidy
final = lastday_filtered.tail(1)
final.T
```

```python
# We only care about the percentages so make a variable that's a list with 
# all the columns that have the string 'pct' in them
pct_columns = list(df.columns[df.columns.str.contains('pct')])
final[pct_columns]
```

```python
#Tidy, make an index for the x-axis buckets
plot_data = final[pct_columns].T
plot_data.index = range(1, 11)
plot_data
```

```python
#Just the same as the fake data but use the actual data just created in
# plot_data
plt.style.use('fivethirtyeight')
fig = plt.figure()


#Set the color of the bars  and their width
ax = plot_data.plot.bar(color='#ef7030', width=0.9)

#set the spacing and number of increments on the y axis
#set the background color of the chart area
ax.set(yticks=range(0,50,10), facecolor='#fcfcfc')

#set the background color of the border 
fig.patch.set_facecolor('#fcfcfc')
#Label the axes and set to bold
plt.ylabel('Percent of total votes', fontsize=9, fontweight='bold')
plt.xlabel('Rating', fontsize=9, fontweight='bold')

#Set the title and give the coordinates for placing the title
ax.text(x=-2, y=46, s="'An Inconvenient Sequel: Truth To Power' is divisive",
       fontsize=12, fontweight='bold')

#Set the sub-title and give the coordinates for placing the title
ax.text(x=-2, y=43, s= 'IMDb ratings for the film as of Aug. 29',
       fontsize=11)

#The numbers on the x-axis were original on their sides. This rotates them
#upright
ax.tick_params(labelrotation=0);
```
