```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
from IPython.display import display, Image
url = 'https://fivethirtyeight.com/wp-content/uploads/2017/09/mehtahickey-inconvenient-0830-1.png'
example = Image(url=url, width=400)

display(example)
```

```python
plt.style.use('fivethirtyeight')
fake = pd.Series([38,3,2,1,2,4,6,5,5,33],
                 index=range(1,11))
fake.plot.bar(color='#ed713a', width=0.9);
```

```python
fig = plt.figure(facecolor='white', figsize=(5,4))

ax = fake.plot.bar(color='#ed713a', width=0.9)
ax.set(facecolor='white', frame_on=False)

plt.xlabel('Rating',fontweight='bold', fontsize=9)
plt.ylabel('percent of total votes', fontweight='bold', fontsize=9)

plt.title('`An Inconvenient Sequel: Truth To Power` is divise',
         fontsize=14,
         loc='left',
         x=-0.1,
         y=1.1, fontweight='bold',
          color='#212121')

plt.text(x=-1.7, y=fake.max() + 4, s='IMDb ratings for the film as of Aug. 29', fontsize=11)

plt.xticks(rotation=0, color='#a7a7a7', fontsize=9)
plt.yticks(range(0,50,10), labels=[f'{i}' if i!=40 else f'{i}%' for i in range(0,50,10)], color='#a7a7a7', fontsize=9);
```
